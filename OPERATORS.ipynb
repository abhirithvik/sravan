```python
# operators

# 1) Arithematic Operators
a = 7
b = 6
c = 10
d = 1
e = 2
f = [1,2,3]
g = '1'
h = '123'

print(a*b) # ("*" operator used for mult two values)
print(a-b) # ("-" used for sub two values)
print(a/b) # ("/" used for dividing two values)
print(a//b) # ("//" used for floor division of two values)
print(a**b) # ("**" used for exponent of two values)
print(a%b) # ("%" used for modulos of two values)


# 2) Comparison Operators

print(a>b) # (">" checking 'a' is greater then 'b' or not)
print(a==b) # ("==" checking 'a' is equal to 'b' or not)
print(a!=b) # ("!=" checking 'a'is not equal to 'b' or not)
print(a>=b) # (">=" Checking whether 'a' is greater than 'b' or equal to 'b')
print(a<=b) # ("<="Checksing wether 'a' is less than 'b' or equal to 'b')


# 3) Identity Operators

print(a is b) # ("is" checking wether 'a' and 'b' having same address location)
print(a is not b) # ("is not" checking 'a' and 'b' having different address location or not)


# 4) Logical Operators

print(a>b and b<c) # ("and" Displays boolean value based on logical and operation)
print(a<b or c>a) # (" or" Displays boolean value based on logical and operation)
print(not(a>b and a<c)) # (" not" it reverses the result)


# 5) Bitwise Operator

print(a&b) #((AND"&") operation bitwise gives output by performing)
print(a | b) #((OR"|") operation bitwise gives output by performing)
print(a^b) #(EXOR"^") operation bitwise gives output by performing)
print(d>>e) #((RIGHT SHIFT">>") operation bitwise gives output by performing)
print(~a,~b) #((NOT"~") operation bitwise gives output by performing)
print(d<<e) #((LEFT SHIFT"<<") operation bitwise gives output by performing)


# 6) membership operator

print(g in h) # ( checking 'g' is in 'h' or not)
print(d not in f) #(checking 'd' is not in 'f'or not)


# 7) assignment operator

a=b
print(a) # (a value is assigned to a)
a+=b
print(a) # (increments a value by b)
a-=b
print(a) # (decrements a value by b) 
a*=b
print(a) # (Multiplies a value by b and stores in a)
a/=b
print(a) # (divides a value by b and stores in a)
a//=b
print(a) # (floor divides a value by b and stores in a)
a%=b
print(a) # (modulos a value by b and stores in a)
a=8
a&=b
print(a) # (performs (&) with a value by b and stores in a)
a|=b
print(a) # (performs (|) with a value by b and stores in a)
a^=b
print(a) # (performs (^) with a value by b and stores in a)
a<<=b
print(a) # (performs (<<) with a value by b and stores in a)


```
